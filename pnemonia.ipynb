```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())

model.add(Dense(128,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_gen=ImageDataGenerator(rescale=1./255)
test_gen=ImageDataGenerator(rescale=1./255)
```

```python
train_data = train_gen.flow_from_directory(
    r"C:\Users\rijin\Downloads\archive (2)\chest_xray\train",
    target_size=(150,150),
    batch_size=32,
    class_mode="binary"
)
```

```python
test_data = test_gen.flow_from_directory(
    r"C:\Users\rijin\Downloads\archive (2)\chest_xray\test",
    target_size=(150,150),
    batch_size=32,
    class_mode="binary"
)
```

```python
history=model.fit(train_data,epochs=10,validation_data=test_data)
```

```python
y_pred=model.predict(test_data)
y_pred
```

```python
import numpy as np

y_pred_class = (y_pred > 0.5).astype(int)
y_pred_class
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(test_data.classes, y_pred_class))
print(classification_report(test_data.classes, y_pred_class))
```
